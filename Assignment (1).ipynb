```python
 # 7) --------------Program---------------

English = int(input("Enter Your English Marks:"))
Mathematics = int(input("Enter Your Mathematics Marks:"))
Computer = int(input("Enter Your Computer Marks:"))
Chemistry = int(input("Enter Your Chemistry Marks:"))
Physics = int(input("Enter Your Physics Marks:"))

obtainmarks = English + Mathematics + Computer + Chemistry + Physics
print("Your Obatin Marks is:",obtainmarks)

perc = obtainmarks /500 * 100
print("Your Precentage is:",perc)

if perc >=80 and perc <=100:
    print("Your Grade is: A1 ")
elif perc >=70 and perc <=80:
    print("Your Grade is: A ")
elif perc >=60 and perc <=70:
    print("Your Grade is: B")
elif perc >=50 and perc <=60:
    print("Your Grade is: C")
elif perc >=40 and perc <=50:
    print("Your Grade is: D")
else:
    print("Fail")
```
